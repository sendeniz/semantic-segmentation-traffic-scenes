```python
def trans_out (h, s, p_in, d, k, p_out):
  res = ( ((h-1)*s) - 2*p_in + (d*(k-1)) + p_out + 1 )
  print (res)
trans_out(h=512, s=2, p_in= 0, p_out = 0, k = 2, d = 1 )

trans_out(h=512, s=2, p_in= 1, p_out = 0, k = 4, d = 1 )
```
